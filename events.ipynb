```python
start = datetime(2020,1,1)

nomenclature = {'Frequency': ('frequency', 'Hz'), 
                
                'Positive Seq Volt Mag': ('pos_Vmag', 'volts'),
                'Positive Seq I Mag': ('pos_Imag', 'amps'),
                'Positive Seq V Ang': ('pos_Vang', 'deg'),
                'Positive Seq I Ang': ('pos_Iang', 'deg'),
                
                'Phase A Volt Mag': ('Ph1_Vmag', 'volts'),
                'Phase B Volt Mag': ('Ph2_Vmag', 'volts'),
                'Phase C Volt Mag': ('Ph3_Vmag', 'volts'),
                
                'Phase A Volt Ang': ('Ph1_Vang', 'deg'),
                'Phase B Volt Ang': ('Ph2_Vang', 'deg'),
                'Phase C Volt Ang': ('Ph3_Vang', 'deg'),
                'PhaseC Volt Ang': ('Ph3_Vang', 'deg'),
                
                'Phase A Amp Mag': ('Ph1_Imag', 'amps'),
                'Phase B Amp Mag': ('Ph2_Imag', 'amps'),
                'Phase C Amp Mag': ('Ph3_Imag', 'amps'),
                
                'Phase A Amp Ang': ('Ph1_Iang', 'deg'),
                'Phase B Amp Ang': ('Ph2_Iang', 'deg'),
                'Phase C Amp Ang': ('Ph3_Iang', 'deg'),
                
                'P (in MW)': ('P', 'megawatts'),
                'Q (in MVAR)': ('Q', 'MVAR'),
                
                'Event Time': ('time', None),
                'Time': ('time', None),
                'Unnamed: 20': (None, None)

               }
                
           
def unit_conversions(df, units_dict):
    for col in units_dict.keys():
        if col == None:
            pass
        elif 'volt_mag' in col:
            units_dict[col] = 'kilovolts'
            df[col] /= 1e3
    
    return df, units_dict
                
```

```python
timestamps = [start + timedelta(seconds=t) for t in df['time']]
ns_timestamps = [btrdb.utils.timez.to_nanoseconds(t) for t in timestamps]
```

```python
t0 = datetime.now()
streams = 0
points = 0
for e in pmus.keys():
        
    for pmu in pmus[e]:
        if pmu.split('.')[1] == 'pdf':
            continue
        df = pd.read_csv(os.path.join(path, e, pmu))
        
        # look up units and new column names using specified nomenclature
        names, units = zip(*[nomenclature[key] for key in df.keys()])
        units_dict = dict(zip(names, units))
        
        # reassign column names
        df = df.rename(columns=dict(zip(df.keys(), names)))
        
#         # convert timestamps to ns
#         timestamps = [start + timedelta(seconds=t) for t in df['time'].tolist()]
#         ns_timestamps = [btrdb.utils.timez.to_nanoseconds(t) for t in timestamps]
        
        # strip pmu name of unnecessary info
        pmu_name = [val for val in pmu.split(' ') if 'station' in val.lower()][0]
        print(pmu_name)
        
        df, units_dict = unit_conversions(df, units_dict)
        
        collection = "lndunn/events/%s/%s"%(e, pmu_name)    
        
        for col in df.keys():
            # skip time and flag errors
            if (col == None) or (col in ['time','P','Q']):
                continue
            elif col not in names or col not in units_dict.keys():
                print('unknown nomenclature for stream \n\t pmu=%s\n\tname=%s'%(pmu_name, col))
                continue
                
            # skip streams we've already added
            if col in [s.name for s in db.streams_in_collection(collection)]:
                continue
                
            # create a new stream
            stream = db.create(uuid=uuid.uuid4(),
                               collection=collection,
                               tags={"name": col, "unit": units_dict[col]})
            
            # insert values into the new stream
            stream.insert(list(zip(ns_timestamps, df[col].tolist())))
            
            streams += 1
            points += len(df)
            
        print('Added %i streams to %s'%(len(db.streams_in_collection('lndunn/transmission_events/%s/%s'%(e, pmu_name))), e))
    print (e)
    
print(streams, points)
print(datetime.now()-t0)
```

```python
def remove_streams(collection, stream_names=None):
    if stream_names == None:
        streams = db.streams_in_collection(collection)
        for s in streams:
            s.obliterate()
            
```

```python
start = datetime(2020,1,1)

def get_units(col_name):
    if col_name == 'Frequency':
        return 'Hz'
    if 'Volt Mag' in col_name:
        return 'Volts'
    if 'Ang' in col_name:
        return 'deg'
    if 'Amp Mag' in col_name:
        return 'Amps'
    if '(in MW)' in col_name:
        return 'MW'
    if '(in MVAR)' in col_name:
        return 'MVAR'
    
renames = {'Event Time': 'Time', 'PhaseC Volt Ang': 'Phase C Volt Ang'}

for e in pmus.keys():
    for pmu in pmus[e]:
        if pmu.split('.')[1] == 'pdf':
            continue
        df = pd.read_csv(os.path.join(path, e, pmu))
        
        names, units = zip(*[nomenclature[key] for key in df.keys()])
        
        renames = dict(zip(df.keys(), names))
        df = df.rename(columns=renames).set_index('time')
        
        for col in df.keys():
            if col == 'time' or col:
                print ('skip', col)
            else:
                pass
            
            
        pmu_name = pmu.split(' ')[2]
        timestamps = [start + timedelta(seconds=t) for t in df['Time'].tolist()]

        collection = "lndunn/transmission_events/%s/%s"%(e, pmu_name)    

        for col_name in df.keys():
            if col_name == 'Time' or 'Unnamed' in col_name:
                continue
            else:
                strm_name = col_name.replace(' ','_')
                unit = get_units(col_name)
                if strm_name in [s.name for s in db.streams_in_collection(collection)]:
                    pass
                else:
                    stream = db.create(uuid=uuid.uuid4(),
                                       collection=collection,
                                       tags={"name": strm_name, "unit": unit}
                                      )
                    
                    ns_timestamps = [btrdb.utils.timez.to_nanoseconds(t) for t in timestamps]
                    vals = list(zip(ns_timestamps, df[col_name].tolist()))
                    stream.insert(vals)
        print('Added %i streams to %s'%(len(db.streams_in_collection('lndunn/transmission_events/%s/'%(e))), e))
    print (e)
```
